### 🔎 Kiểm tra và Giải nén Dữ liệu
Đảm bảo file zip của bạn chứa các file `.npy` trực tiếp bên trong hoặc trong thư mục `keypoints/`.

```python
PATH_TO_ZIP = "/content/drive/MyDrive/LuminaSign/keypoints.zip"

import os
if os.path.exists(PATH_TO_ZIP):
    # Xóa thư mục cũ nếu có để tránh lồng thư mục
    !rm -rf keypoints
    # Giải nén
    !unzip -q {PATH_TO_ZIP} -d ./
    
    # --- KIỂM TRA CẤU TRÚC SAU GIẢI NÉN ---
    if os.path.exists('keypoints/keypoints'):
        print("⚠️ Phát hiện thư mục lồng nhau, đang sửa...")
        !mv keypoints/keypoints/* ./keypoints/
        !rm -rf keypoints/keypoints
    
    num_files = len([f for f in os.listdir('keypoints') if f.endswith('.npy')]) if os.path.exists('keypoints') else 0
    print(f"✅ Đã tìm thấy {num_files} file keypoints (.npy) trong thư mục training/keypoints/")
else:
    print("❌ LỖI: Không tìm thấy file zip trên Drive tại: " + PATH_TO_ZIP)
```
